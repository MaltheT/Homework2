```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from pandas import *
from mpl_toolkits.axes_grid1 import Divider, Size
from mpl_toolkits import mplot3d
plt.rcParams['figure.figsize'] = [10, 5]


# f1tenth simulator data
x_data_1 = np.genfromtxt('test1/x.csv', delimiter=',')
y_data_1 = np.genfromtxt('test1/y.csv', delimiter=',')
z_data_1 = np.genfromtxt('test1/z.csv', delimiter=',')
distance_1 = np.genfromtxt('test1/distance.csv', delimiter=',')


x_data_2 = np.genfromtxt('test2/x.csv', delimiter=',')
y_data_2 = np.genfromtxt('test2/y.csv', delimiter=',')
z_data_2 = np.genfromtxt('test2/z.csv', delimiter=',')
distance_2 = np.genfromtxt('test2/distance.csv', delimiter=',')


x_data_3 = np.genfromtxt('test3/x.csv', delimiter=',')
y_data_3 = np.genfromtxt('test3/y.csv', delimiter=',')
z_data_3 = np.genfromtxt('test3/z.csv', delimiter=',')
distance_3 = np.genfromtxt('test3/distance.csv', delimiter=',')


#simulation time
simtime = 15
```

```python
#plot distance from drone to traget pose

t_1 = np.linspace(0, simtime, len(distance_1))
t_2 = np.linspace(0, simtime, len(distance_2))
t_2 = np.linspace(0, simtime, len(distance_3))


t_r = np.linspace(0, simtime, len(distance_1))
reference = np.zeros(len(t_r))


plt.plot(t_1, distance_1, color='blue', label='Kp = 0.5, Ki = 0.0, Kd = 0.1')
plt.plot(t_2, distance_2, color='orange', label='Kp = 0.5, Ki = 0.001, Kd = 0.1')
plt.plot(t_2, distance_3, color='green', label='Kp = 0.5, Ki = 0.01, Kd = 0.1')
plt.plot(t_r, reference, color='black', linestyle='dashed',label='refence')

plt.xlabel('time')
plt.ylabel('distance to goal')
plt.legend()
```
